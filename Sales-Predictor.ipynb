```python
import pandas as pd

data = pd.read_csv('TV_Advertising_vs_Sales_Dataset.csv')
print(data.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['TV'], data['Sales'])
plt.title('TV Advertising Spend VS Sales')
plt.xlabel('TV Advertising Spend (in $)')
plt.ylabel('Sales (in $)')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

x = data[['TV']] #Independent variable
y = data['Sales'] #Dependent variable

model = LinearRegression()
model.fit(x,y)

print("Slope (coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
new_tv_spend = [[50], [100], [200]]
predicted_sales = model.predict(new_tv_spend)

print("Predicted Sales:", predicted_sales)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(x)

mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Mean Squared error:", mse)
print("R2 Score:", r2)
```

```python
plt.scatter(data['TV'], data['Sales'], label="Data points")
plt.plot(data['TV'], y_pred, color='red', label="Regression Line")
plt.title("TV Advertising Spend vs Sales with Regression Line")
plt.xlabel("TV Advertising Spend (in $)")
plt.ylabel("Sales (in $)")
plt.legend()
plt.show()
```

```python
residuals = y - y_pred
plt.scatter(y_pred, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals Plot')
plt.xlabel('Predicted Sales')
plt.ylabel('Residuals')
plt.show()
```
